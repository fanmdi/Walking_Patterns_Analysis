```python
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

# Assuming your dataframe is named 'df'
target_variable = 'diff_pre_during'
independent_variables = [ 'Commercial_during', 'Education_during', 'Hospital/Medical_during',
       'Industrial_during', 'Parkland_during', 'Residential_during',
       'Transport_during', 'Water_during',
       'income_person_during', 'population_during','intersection_count_pre',
       'street_node_avg_during', 'num_links_nodes_ratio_during',
       'POI_during']

# Create a new dataframe with only the selected columns
data = diff_pre_during[[target_variable] + independent_variables].copy()

# Drop rows with missing values, if any
data.dropna(inplace=True)

# Separate the independent variables (X) and the target variable (y)
X = data[independent_variables]
y = data[target_variable]

# Standardize the independent variables
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Initialize the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model to the data
rf_model.fit(X, y)

# Get the feature importances
importances = rf_model.feature_importances_

# Create a dataframe with the variable names and their feature importances
importances_df = pd.DataFrame({
    'Variable': independent_variables,
    'Feature Importance': importances
})

# Sort the dataframe by feature importance in descending order
importances_df = importances_df.sort_values(by='Feature Importance', ascending=False)

# Print the dataframe
print(importances_df)
```

Feature importances are provided by the fitted attribute feature_importances_ and they are computed as the mean and standard deviation of accumulation of the impurity decrease within each tree.
Warning: Impurity-based feature importances can be misleading for high cardinality features (many unique values). See Permutation feature importance as an alternative below.

```python
import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.ensemble import RandomForestRegressor
from sklearn.preprocessing import StandardScaler

# Assuming your dataframe is named 'df'
target_variable = 'diff_pre_during'
independent_variables = [ 'Commercial_during', 'Education_during', 'Hospital/Medical_during',
       'Industrial_during', 'Parkland_during', 'Residential_during',
       'Transport_during', 'Water_during',
       'income_person_during', 'population_during','intersection_count_pre',
       'street_node_avg_during', 'num_links_nodes_ratio_during',
       'POI_during']

# Create a new dataframe with only the selected columns
data = diff_pre_during[[target_variable] + independent_variables].copy()

# Drop rows with missing values, if any
data.dropna(inplace=True)

# Separate the independent variables (X) and the target variable (y)
X = data[independent_variables]
y = data[target_variable]

# Standardize the independent variables
scaler = StandardScaler()
X = scaler.fit_transform(X)

# Initialize the Random Forest model
rf_model = RandomForestRegressor(n_estimators=100, random_state=42)

# Fit the model to the data
rf_model.fit(X, y)

# Compute feature importances using MDI method
start_time = time.time()
importances = rf_model.feature_importances_
std = np.std([tree.feature_importances_ for tree in rf_model.estimators_], axis=0)
elapsed_time = time.time() - start_time

print(f"Elapsed time to compute the importances: {elapsed_time:.3f} seconds")

# Update the feature names with the desired variable names
feature_names = independent_variables.copy()
feature_names[feature_names.index('income_person_during')] = 'Weekly Person Income'

# Create a Series with feature importances and corresponding feature names
forest_importances = pd.Series(importances, index=feature_names)

# Sort feature importances in descending order
forest_importances = forest_importances.sort_values(ascending=False)

# Remove '_during' from the variable names
forest_importances.index = forest_importances.index.str.replace('_during', '')

# Update the variable names in the index of forest_importances
new_variable_names = {
    'Industrial': '% of industrial land use',
    'population': 'Population',
    'Education': '% of education land use',
    'Residential': '% of residential land use',
    'Parkland': '% of parkland land use',
    'Commercial': '% of commercial land use',
    'income_person': 'Weekly Person Income',
    'Water': '% of water land use',
    'Transport': '% of transport land use',
    'Hospital/Medical': '% of hospital/medical land use',
    'edge_density_km': 'Edge density',
    'num_links_nodes_ratio': 'Number of links/Number of nodes',
    'street_node_avg': 'Average street nodes',
    'intersection_count_pre': 'Intersection count'
}

# Replace the variable names in the index with the new names
forest_importances.index = forest_importances.index.to_series().replace(new_variable_names, regex=True)

# Convert std to a pandas Series with the same index as forest_importances
std_series = pd.Series(std, index=forest_importances.index)

# Set the font to Calibri
plt.rcParams['font.family'] = 'Calibri'

# Set the style with grids for both x and y
sns.set_style('whitegrid')

# Create a color map for the bars based on the feature importances (higher importance, redder color)
color_map = sns.color_palette("coolwarm", n_colors=len(forest_importances))
color_map = color_map[::-1]

# Plot feature importances using a bar chart with custom colors and standard deviation lines
fig, ax = plt.subplots(figsize=(10, 6))  # Increase the figure size

# Use barh for horizontal bars
sns.barplot(x=forest_importances.values, y=forest_importances.index, palette=color_map, ax=ax, ci='sd')

# Add error bars (standard deviation lines)
for index, value in enumerate(forest_importances):
    ax.errorbar(value, index, xerr=std_series[index], color='black', capsize=5)

ax.set_xlabel("Feature importance scores", fontsize=20)
ax.set_ylabel("Selected Variable in Pre-During COVID", fontsize=20)

# Add x-axis limit from -0.4 to 0.8
plt.xlim(-0.4, 0.8)

# Change the font size of variable names on the y-axis
plt.yticks(fontsize=20)

# Change the font size of variable names on the x-axis
plt.xticks(fontsize=20)

# Add grid lines in the background
#plt.grid(axis='both', linestyle='--')
plt.grid(False)
fig.tight_layout()

# Save the plot as a JPG image
plt.savefig('RandomForest_coefficients_pre_plot.pdf', dpi=300, bbox_inches='tight')
plt.show()
```
